

### **1- What is Simple Linear Regression?**

Simple Linear Regression is a statistical method used to model the relationship between **one independent variable (X)** and **one dependent variable (Y)** using a straight line.
The model predicts Y based on X.

The equation is:
"Y = mX + c"

Where **m** is the slope and **c** is the intercept.

---

### **2- What are the key assumptions of Simple Linear Regression?**

The key assumptions are:

1. **Linearity** – Relationship between X and Y is linear.
2. **Independence** – Observations are independent.
3. **Homoscedasticity** – Constant variance of residuals.
4. **Normality** – Residuals are normally distributed.
5. **No significant outliers**.

---

### **3- What does the coefficient m represent in the equation Y=mX+c?**

The coefficient **m** represents the **slope of the regression line**.

It indicates **how much the dependent variable (Y) changes for a one-unit increase in the independent variable (X).**

---

### **4- What does the intercept c represent in the equation Y=mX+c?**

The intercept **c** represents the **value of Y when X = 0**.

It shows where the regression line **intersects the Y-axis**.

---

### **5- How do we calculate the slope m in Simple Linear Regression?**

The slope is calculated using:

m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}

This formula measures how **X and Y vary together**.

---

### **6- What is the purpose of the least squares method in Simple Linear Regression?**

The **least squares method** finds the best-fitting regression line by **minimizing the sum of squared differences between actual and predicted values**.

Purpose:

* Reduce prediction error
* Find the best regression line

---

### **7- How is the coefficient of determination (R²) interpreted in Simple Linear Regression?**

The coefficient of determination is:

R^2 = 1 - \frac{SS_{res}}{SS_{tot}}

Interpretation:

* **R² = 1** → Perfect prediction
* **R² = 0** → Model explains no variance
* Higher R² means the model explains more variability in Y.

---

### **8- What is Multiple Linear Regression?**

Multiple Linear Regression is a statistical technique used to model the relationship between **one dependent variable and two or more independent variables**.

Example equation:

[
Y = b_0 + b_1X_1 + b_2X_2 + ... + b_nX_n
]

---

### **9- What is the main difference between Simple and Multiple Linear Regression?**

| Simple Linear Regression          | Multiple Linear Regression                 |
| --------------------------------- | ------------------------------------------ |
| Uses **one independent variable** | Uses **two or more independent variables** |
| Simpler model                     | More complex model                         |

---

### **10- What are the key assumptions of Multiple Linear Regression?**

1. Linearity
2. Independence of errors
3. Homoscedasticity
4. Normal distribution of residuals
5. No multicollinearity between independent variables.

---

### **11- What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**

Heteroscedasticity occurs when the **variance of residuals is not constant across all levels of independent variables**.

Effects:

* Biased standard errors
* Incorrect hypothesis testing
* Less reliable predictions.

---

### **12- How can you improve a Multiple Linear Regression model with high multicollinearity?**

Methods include:

* Removing highly correlated variables
* Using **Principal Component Analysis (PCA)**
* Using **Ridge or Lasso Regression**
* Combining correlated variables.

---

### **13- What are some common techniques for transforming categorical variables for use in regression models?**

Common techniques:

* **One-hot encoding**
* **Label encoding**
* **Dummy variables**
* **Binary encoding**

---

### **14- What is the role of interaction terms in Multiple Linear Regression?**

Interaction terms show how **two variables together influence the dependent variable**.

Example:

[
Y = b_0 + b_1X_1 + b_2X_2 + b_3(X_1X_2)
]

They help model **combined effects between variables**.

---

### **15- How can the interpretation of intercept differ between Simple and Multiple Linear Regression?**

* In **Simple Regression**, intercept = value of Y when X = 0.
* In **Multiple Regression**, intercept = value of Y when **all independent variables are 0**.

---

### **16- What is the significance of the slope in regression analysis, and how does it affect predictions?**

The slope indicates the **rate of change between variables**.

* Positive slope → Y increases as X increases
* Negative slope → Y decreases as X increases.

---

### **17- How does the intercept in a regression model provide context for the relationship between variables?**

The intercept provides the **baseline value of the dependent variable when independent variables are zero**.
It helps interpret the starting point of the regression line.

---

### **18- What are the limitations of using R² as a sole measure of model performance?**

Limitations:

* Does not detect **overfitting**
* Always increases when variables are added
* Does not measure prediction accuracy.

---

### **19- How would you interpret a large standard error for a regression coefficient?**

A large standard error indicates that the **coefficient estimate is not precise** and may not be statistically significant.

---

### **20- How can heteroscedasticity be identified in residual plots, and why is it important to address it?**

Identification:

* Residual plot shows **funnel or cone shape pattern**.

Importance:

* It affects **statistical tests and confidence intervals**, making results unreliable.

---

### **21- What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?**

This indicates **too many irrelevant variables in the model**.

Adjusted R² penalizes unnecessary predictors.

---

### **22- Why is it important to scale variables in Multiple Linear Regression?**

Scaling helps:

* Improve **model stability**
* Reduce **multicollinearity**
* Improve performance in **regularized models (Ridge, Lasso)**.

---

### **23- What is polynomial regression?**

Polynomial regression is a type of regression where the relationship between variables is modeled as an **nth-degree polynomial** instead of a straight line.

---

### **24- How does polynomial regression differ from linear regression?**

| Linear Regression          | Polynomial Regression |
| -------------------------- | --------------------- |
| Straight line relationship | Curved relationship   |
| Equation degree = 1        | Degree > 1            |

---

### **25- When is polynomial regression used?**

Polynomial regression is used when the **relationship between variables is nonlinear but can be approximated by a polynomial curve**.

Example: growth patterns, market trends.

---

### **26- What is the general equation for polynomial regression?**

[
Y = a + b_1X + b_2X^2 + b_3X^3 + ... + b_nX^n
]

---

### **27- Can polynomial regression be applied to multiple variables?**

Yes. Polynomial regression can be extended to **multiple independent variables**, forming **multivariate polynomial regression**.

---

### **28- What are the limitations of polynomial regression?**

* Risk of **overfitting**
* Sensitive to **outliers**
* High-degree polynomials may behave unpredictably outside the data range.

---

### **29- What methods can be used to evaluate model fit when selecting the degree of a polynomial?**

Methods include:

* Cross-validation
* Adjusted R²
* Mean Squared Error (MSE)
* AIC/BIC criteria.

---

### **30- Why is visualization important in polynomial regression?**

Visualization helps:

* Understand the **curve fit**
* Detect **overfitting or underfitting**
* Compare polynomial degrees visually.

---

### **31- How is polynomial regression implemented in Python?**

Example in Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([1,4,9,16,25])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

y_pred = model.predict(X_poly)

plt.scatter(X, y)
plt.plot(X, y_pred)
plt.show()
```
